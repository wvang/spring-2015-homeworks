# Analysis of Armed Conflicts in Africa

Using data from the ACLED project (http://www.acleddata.com/), we try to analyze some trends that occur in violent events in  Africa. The dataset is a list of events that includes the people who were involved in the event (ACTOR1, ACTOR2, etc), the geographic location, the fatality count of the event, the type of interaction, the event type, and many other fields. More information can be found in the dataset description. (http://www.acleddata.com/wp­content/uploads/201 5/01 /ACLED_Codebook_201 5.pdf).

## Hypothesis 1: Actors tend to have similar fatality counts across their events

It's reasonable to believe that groups tend to act in the same way across events. Peaceful protesters would always have no casualty count, while military forces would have much higher. Fatality count is a major decision for groups, and definately defines how their preferred behavior simply because of the magnitude of the decision. We will try to show this hypothesis is true by taking a look at the standard deviations of the actor's fatality counts.

```python
fatalities_mean = []
fatalities_std = []
for actor in data['ACTOR1'].unique():
    actor_data = data[data['ACTOR1'] == actor]
    actor_fatalities_std = actor_data['FATALITIES'].std()
    actor_fatalities_mean = actor_data['FATALITIES'].mean()
    if not math.isnan(actor_fatalities_std) and not math.isnan(actor_fatalities_mean):
        fatalities_std.append(actor_fatalities_std)
        fatalities_mean.append(actor_fatalities_mean)
```

```python
fig, ax = plt.subplots()
ax.scatter(fatalities_mean,fatalities_std)
plt.xlabel('Mean of actor')
plt.ylabel('STD of actor')
plt.plot()
print 'Mean of std values:',np.mean(fatalities_std)
print 'Median of std values:', np.median(fatalities_std)
print 'Standard deviation of std values:', np.std(fatalities_std)
```

We take the mean of the fatalities from events for ACTOR1 and plot it against the standard deviation of ACTOR1. We can see here that a larger mean doesn't necessarily constitute a larger std dev- there are just as many large mean values with small std dev as they are large std dev. The standard deviations seem to cluster closer to the origin. Next we will remove the outliers to take a closer look at the origin.

```python
fatalities2_mean = []
fatalities2_std = []
for actor in data['ACTOR1'].unique():
    actor_data = data[data['ACTOR1'] == actor]
    actor_fatalities_std = actor_data['FATALITIES'].std()
    actor_fatalities_mean = actor_data['FATALITIES'].mean()
    if not math.isnan(actor_fatalities_std) and not math.isnan(actor_fatalities_mean):
        if actor_fatalities_std < 2000 and actor_fatalities_mean < 100:
            fatalities2_std.append(actor_fatalities_std)
            fatalities2_mean.append(actor_fatalities_mean)
```

```python
fig, ax = plt.subplots()
ax.scatter(fatalities2_mean,fatalities2_std)
plt.xlabel('Mean of actor')
plt.ylabel('STD of actor')
plt.plot()
```

Closer to the origin, we can see that in general as the mean increases, so does the standard deviation.

```python
plt.hist(fatalities2_std, bins=20)
plt.xlabel('STD of actor')
plt.plot()
```

Above is a histogram of the STD values for ACTOR1. As you can see, a overwhelming percentage of actors tend to have low standard deviation compared to others. This would imply that our hypothesis that actors tend to have similar fatality counts across events is true, with the caveat that as actors have higher average fatality counts in events they commit, their standard deviation grows. In other words, peaceful groups tend to stay peaceful, but non-peaceful groups tend to vary in their fatality counts.

## Hypothesis 2: Events where a militaristic and non-militaristic group are involved tend to have more fatalities

Since one side tends to be peaceful, this would imply they either lack weapons or the ability/desire to take lives. If pitted against a militaristic group, their lesser combat ability would imply more casualties in general. 

First, we need to define a non-militaristic group and militaristic group are. A non-militaristic group is an actor that has taken very few lives. We'll define it as groups that take on average less than 5 lives per event. All other groups can be considered militaristic.

```python
mean_aggressor = data[data['EVENT_TYPE'] == 'Violence against civilians']['FATALITIES'].mean()
mean_non = data[data['EVENT_TYPE'] != 'Violence against civilians']['FATALITIES'].mean()
print float(mean_aggressor)/float(mean_non)
```

A simple ratio check between the event type 'Violence against civilians''s mean fatalities vs non-'Violence against civilians' would suggest that there are in fact more fatalities in militaristic group vs militaristic group, but we can go more indepth then that.

```python
fatalities_mean = []
fatalities_std = []
actor_militaristic = {}
data['ACTOR1AF'] = 0
data['ACTOR2AF'] = 0
for actor in data['ACTOR1'].unique():
    actor_data = data[data['ACTOR1'] == actor]
    actor_fatalities_mean = actor_data['FATALITIES'].mean()
    if math.isnan(actor_fatalities_mean):
        actor_fatalities_mean = 0
    data.ix[data.ACTOR1==actor,'ACTOR1AF'] = actor_fatalities_mean
    data.ix[data.ACTOR2==actor,'ACTOR2AF'] = actor_fatalities_mean
```

```python
cutoff = 0
cutoffs = []
ratios = []
while (cutoff <= 220):
    events = data[(data['ACTOR1AF'] > cutoff) & (data['ACTOR2AF']  <= cutoff)]
    nevents = data[(data['ACTOR1AF'] <= cutoff) | (data['ACTOR2AF']  > cutoff)]
    ratio = float(events['FATALITIES'].mean())/float(nevents['FATALITIES'].mean())
    cutoffs.append(cutoff)
    ratios.append(ratio)
    cutoff+=10
plt.plot(cutoffs,ratios)
plt.xlabel('Cutoff for violent/non-violent')
plt.ylabel('Ratio')
```

As you can see, the ratio for the fatalities for violent against non-violent type events vs all others is always above 1 and is large event for small cutoffs. Therefore we can confirm that violent vs non-violent events have much more fatlities than other event types. 

## Hypothesis 3: Medium to low casualty events tend to occur in the same geographic location, while higher casualty events tend to be more geographically widespread.

Due to their costly nature, one would assume that high casualty events would occur less often. Can the same be assumed of their geographic nature. High casualty events would seem likely to be outliers rather than the norm. If repeated high casualty events occured in the same area, the stability and desire to live in that area would decrease, and less people would go to that area and thus less casualties.
