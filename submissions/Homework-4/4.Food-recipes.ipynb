# Clustering food recipes from around the world

In this assignment, you are expected to hierarchically cluster food recipes. We will be using YY Ahn's dataset "[Recipe datasets with cuisines](http://yongyeol.com/data/scirep-cuisines-detail.zip)", which for each recipe provides a list of ingredients and the country of origin, scraped from popular recipe websites. ```python

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition.truncated_svd import TruncatedSVD
import scipy.cluster.hierarchy as hcluster
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans 
from scipy.sparse import * 
import pandas as pd
import sklearn.datasets as sk_data
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

'''HYPOTHESIS: Recipes from the same origin are similar.
How do we prove this? First, we take out the origin data from the recipes
and cluster them solely on their ingredients. Then, after we've clustered them,
we tally up the number of recipes in each cluster for each origin (i.e. america could have 
5 in cluster 1, 10 in cluster 2, etc.). By looking at the percentage of requests in a cluster
in a particular origin, we can try and ascertain whether recipes from the same origin are similar.
'''

recipes = []
#first loop through and store all the recipes as text
with open('allr_recipes.txt') as f:
    for line in f:
        vector = line.strip('\n').split("\t")
        vector = vector[1:]
        vector = ' '.join(vector)
        recipes.append(vector)
with open('epic_recipes.txt') as f:
    for line in f:
        vector = line.strip('\n').split("\t")
        vector = vector[1:]
        vector = ' '.join(vector)
        recipes.append(vector)
with open('menu_recipes.txt') as f:
    for line in f:
        vector = line.strip('\n').split("\t")
        vector = vector[1:]
        vector = ' '.join(vector)
        recipes.append(vector)
#then use the vecotrizer to convert the text into categories
#turn them into a count matrix (which will be binary)
#and then transform that into a normalized tf-idf represemtation
#we normalize it because there are certain ingredients i.e. eggs, flour, water
#that are used universally and don't define the uniqueness of a recipe
#therefore we normalize by scaling down the impact of ingredient that occur very
#frequently and  that are hence empirically less informative 
#than features that occur in a small fraction of the recipes
vectorizer = TfidfVectorizer(min_df=1)
X = vectorizer.fit_transform(recipes)
#next we reduce the data
#there are 383 unique ingredients, giving us a 50000 x 383 or so matrix. 
#we need to reduce it to be able to compute clustering without running of memory
pca = TruncatedSVD(n_components=10)                               
X_reduced_train = pca.fit_transform(X)
#even after reducing, DBSCAN and Hierarchical clutering didn't run well because of memory
#therefore I decided to go with K-Means. Below I graph the error, and after looking at the graph
#I decided upon 10 clusters to minimize clusters and error.
error = np.zeros(20)
error[0] = 0;
for k in range(1,20):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(X_reduced_train)
    error[k] = kmeans.inertia_
plt.plot(range(1,len(error)),error[1:])
plt.xlabel('Number of clusters')
plt.ylabel('Error')

kmeans = KMeans(init='k-means++', n_clusters=10, n_init=10)
kmeans.fit_predict(X_reduced_train)

origins = []
#get all the origins now
with open('allr_recipes.txt') as f:
    for line in f:
        vector = line.strip('\n').split("\t")
        vector = vector[0].lower()
        origins.append(vector)
with open('epic_recipes.txt') as f:
    for line in f:
        vector = line.strip('\n').split("\t")
        vector = vector[0].lower()
        origins.append(vector)
with open('menu_recipes.txt') as f:
    for line in f:
        vector = line.strip('\n').split("\t")
        vector = vector[0].lower()
        origins.append(vector)

#now we store how many recipes are in each cluster for a particular origin
#we store each origin as a key in a dict, and a numpy array that tracks number
#of recipes in each cluster
clusterMappings = {}
labels = kmeans.labels_
for i in range(len(origins)):
    origin = origins[i]
    originMap = clusterMappings.setdefault(origin,np.zeros(10))
    originMap[labels[i]]+= 1
    clusterMappings[origin] = originMap
    

#now we need to convert the quantities in each array into percentages-
#as we want to see how the recipes are clustered as a part of a whole, not a simple quanity
percentageMappings = {}
for k, v in clusterMappings.items():
    pv = v/sum(v)
    percentageMappings[k] = pv
percentageData = pd.DataFrame(percentageMappings).T

#now we create a clustermap, which is like a heatmap but it groups together similar information
#to make it easier to find patterns
#Essentially, it will display how large of a percentage of the origin's recipes are in a cluster
#by the color internsity. 
'''Observe the clustermap below. Things to note:
1) Look at the Asian countries- they all seem to have a high percentage of their
recipes contained within cluster 6. This indicates similar recipes within their own origin as well
as with the other Asian origins.
2) Look at the African/Middle-east contries. They too seem to have a high percentage of their
recipes contained within a single cluster- cluster 5
3) Looking over the Mediteranean area, we see italy, greece, and mediterannean all highly
concentrated in cluter 8
4) What may be more telling is the lack of concentration in other countries,
which all seem to be European and American regions. There may be a myrida of reasons for the
diverse recipes, such as many different ethnic backgrounds in the population.

CONCLUSION:
Certain origins, such as those in the Asian region or the Middle East region, have similar recipes
as exhibited by the high percentage of recipes in a single cluster. Other, such as America and
Europe, do not. 
'''
sns.clustermap(percentageData)
```
